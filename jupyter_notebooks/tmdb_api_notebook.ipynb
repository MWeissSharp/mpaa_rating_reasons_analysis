```python
import requests

import pandas as pd

import json
```

```python
all_mpaa_info = pd.read_csv('../data/clean_mpaa_data.csv')
mpaa_info = all_mpaa_info.drop(columns = 'Unnamed: 0')
mpaa_info.info()
```

```python
mpaa_test = mpaa_info.iloc[:10]
mpaa_test['title_fix'] = mpaa_test['title'].str.endswith(', The')
mpaa_test['title'] = mpaa_test['title'].str.replace(', The', '')

for ind, row in mpaa_test.iterrows():
    if row['title_fix'] == True:
        mpaa_test.loc[ind, 'clean_title'] = 'The ' + row['title']
    else:
         mpaa_test.loc[ind, 'clean_title'] = row['title']
    
mpaa_test['title_fix'] = mpaa_test['title'].str.endswith(', A')
mpaa_test['title'] = mpaa_test['title'].str.replace(', A', '')

for ind, row in mpaa_test.iterrows():
    if row['title_fix'] == True:
        mpaa_test.loc[ind, 'clean_title'] = 'A ' + row['title']
    else:
         mpaa_test.loc[ind, 'clean_title'] = row['clean_title']

mpaa_test = mpaa_test.drop(columns = 'title_fix')
    
mpaa_test
```

```python
with open('mpaa_keys.json') as fi:
    credentials = json.load(fi)
    
api_key = credentials['tmdb_key']
```

```python
endpoint = 'https://api.themoviedb.org/3/search/movie'
```

```python
from tqdm import tqdm
```

```python
mpaa_test['tmdb_res'] = ''
mpaa_test['tmdb_res'] = mpaa_test['tmdb_res'].astype(object)

for ind, row in tqdm(mpaa_test.iterrows()):
    
    params = {
        'api_key' : api_key,
        'query' : row['clean_title']
    }
    
    response = requests.get(endpoint,
                            params=params)
    
    mpaa_test.at[ind, 'tmdb_res'] = response.json()['results']
    mpaa_test.loc[ind, 'tmdb_count'] = len(response.json()['results'])
```
