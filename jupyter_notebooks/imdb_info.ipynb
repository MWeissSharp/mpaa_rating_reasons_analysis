```python
from imdb import Cinemagoer

import pandas as pd
import numpy as np
import pickle

from tqdm import tqdm
```

```python
#read in TMDB data, filter down to just the relevant columns
tmdb = pd.read_pickle('../data/tmdb_results.pkl')
tmdb = tmdb[['imdb_id', 'title']]
tmdb['imdb_id'] = tmdb['imdb_id'].str.replace('tt', '')
```

```python
# Remove rows that do not include an IMDB id
tmdb_imdb = tmdb.loc[~((tmdb['imdb_id'].isna())
                     |
                     (tmdb['imdb_id'] == ''))].reset_index(drop=True)
```

```python
# Initiate Cinemagoer
ia = Cinemagoer()
```

```python
# Iterate through the IMDB id number
local_titles = []
titles = []
genres = []
budgets = []
grosses = []
mpaas = []
ratings = []
votes = []
years = []
production_cos = []
distributors = []

for n in tqdm(range(0, len(tmdb_imdb))):
    
    imdb_id = tmdb_imdb['imdb_id'].iloc[n]
    
    movie = ia.get_movie(imdb_id , info = ['main'])
    
    if 'localized title' in movie:
        loc_title = movie['localized title']
    else:
        loc_title = ''
        
    if 'title' in movie:
        title = movie['title'] 
    else:
        title = ''
        
    if 'genres' in movie:
        genre = movie['genres']
    else:
        genre = ''
    
    if 'certificates' in movie:
        certs = movie['certificates']
        mpaa = [item for item in certs if item.startswith('United States')]
    else:
        mpaa = ''
    
    if 'rating' in movie:
        rating = movie['rating']
    else:
        rating = ''
        
    if 'votes' in movie:
        vote = movie['votes']
    else:
        vote = ''
        
    if 'year' in movie:
        year = movie['year']
    else:
        year = ''
        
    if 'production companies' in movie:
        production = movie['production companies']
    else:
        production = ''
        
    if 'distributors' in movie:
        distributor = movie['distributors']
    else:
        distributor = ''
        
    if 'box office' in movie:
        if 'Budget' in movie['box office']:
            budget = movie['box office']['Budget']
        else:
            budget = ''
        if 'Cumulative Worldwide Gross' in movie['box office']:
            gross = movie['box office']['Cumulative Worldwide Gross']
        else:
            gross = ''
    else:
        budget = ''
        gross = ''
    
    local_titles.append(loc_title)
    titles.append(title)
    genres.append(genre)
    budgets.append(budget)
    grosses.append(gross)
    mpaas.append(mpaa)
    ratings.append(rating)
    votes.append(vote)
    years.append(year)
    production_cos.append(production)
    distributors.append(distributor)
```
